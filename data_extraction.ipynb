```python
import pandas as pd
import os
```

```python
cust_folder = "bigdata2021data/customer_big.parquet/"
tran_folder = "bigdata2021data/transaction_big.parquet/"
```

```python
cust_list = os.listdir(cust_folder)
cust_list.remove('_SUCCESS')
```

```python
def data_extract(folderpath,index_list):
    """index_list give the list of indices [0-199 for customer data]
    (0-121) for transaction data
    folderpath is the path of the folder"""
    filelist = os.listdir(folderpath)
    if '_SUCCESS' in filelist:
        filelist.remove('_SUCCESS')
    
    ind = [int(x[7:10]) for x in filelist]
    file_info = pd.DataFrame([ind, filelist]).transpose()
    file_info.columns = ['file_index', 'filename']
    file_info.set_index('file_index', inplace= True)
    
    filenames = file_info.iloc[index_list]['filename']
    
    output = []
    for i in filenames:
        
        df = pd.read_parquet(folderpath+i)
        output.append(df)
    #return concatenated dataframe
    return pd.concat(output, axis = 0)
    
```
